```python
import matplotlib.pyplot as plt
from hd75732_data import load_hd75732_data

# Load data
time, rv, rv_err = load_hd75732_data()

# Plot
plt.errorbar(time, rv, yerr=rv_err, fmt='o', ecolor='gray', capsize=2)
plt.xlabel("Time (JD)")
plt.ylabel("Radial Velocity (m/s)")
plt.title("HD 75732 Radial Velocity Data")
plt.savefig("hd75732_plot.pdf")  # Save as PDF
plt.show()
```
